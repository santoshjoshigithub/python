## Seaborn : Statistical Data Visualization

**Reference:** https://seaborn.pydata.org/

**Seaborn is a statistical data visualization library for Python which is built on top of Python's Matplotlib library. Here we will see important data visualization which helps us deep dive in our data that will helps us understand it better**

## 1. Import Library

```python
import numpy as np #Linear Algebra
import pandas as pd #data manipulation
import matplotlib.pyplot as plt #base data visualization
import seaborn as sns #data visualization (why sns?)
%matplotlib inline
```

## 2. Collect Sample Data

**Following data set will be primarily used in all our examples**

```python
tips = sns.load_dataset('tips')
sns.set_style("darkgrid")
```

## 3. Distribution Plots

** 3.1 Distribution Plot **

```python
sns.distplot(tips['total_bill'],bins=30)
```

** 3.2 Joint plot **

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

** 3.3 Pair Plot **

```python
sns.pairplot(tips, hue='sex')
```

## 4. Categorical Plots

** 4.1 Box Plot**

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

** 4.2 Count Plot **

```python
sns.countplot(x='sex',data=tips)
```

** 4.3 Factor Plot** 

```python
sns.factorplot(data=tips, x = 'sex', kind = 'count', col='time', hue = 'size', row = 'day')
```
